```python
## Location and distance

bus_stops = pd.read_csv('bus_stops.csv')

MBS = (1.2839454495737048, 103.85937544472274)
busstop_near_MBS = (1.2806844615667559, 103.85507356421047)

## I think with this we should extend radius of end to maybe 600m [10 min walking distance for big attractions]

test = (bus_stops[bus_stops['BusStopCode'] == 28389]['Latitude'].item(), bus_stops[bus_stops['BusStopCode'] == 28389]['Longitude'].item())

## Code for getting distance given matrix and point
def calculate_distance(row, lat, long):
    point = (row['Latitude'], row['Longitude'])
    distance = geopy.distance.geodesic((lat, long), point).m  # Calculate distance in meters
    return distance

## Code for getting closest bus_stop
def closest_bus_stop(lat,long,bus_stops):
    temp = bus_stops.copy()
    temp['dist'] = temp.apply(calculate_distance, args=(lat, long), axis=1)
    closest_bus_stop_idx = temp['dist'].idxmin()
    return temp.loc[closest_bus_stop_idx]['BusStopCode'],temp.loc[closest_bus_stop_idx]['RoadName']
```
